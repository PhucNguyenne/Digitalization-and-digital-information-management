```python
import pandas as pd
import numpy as np

data = {
    "ID": [101,102,103,104,105],
    "Name": ["Quynh Nhu", "Hoang Hai", None, "Phuong Tuan", "Thien An"],
    "Age": [26, 18, 20, None, 19],
    "Salary": [50000, 35000, 65000, 100000, None]
}

df = pd.DataFrame(data)

print(df)
```

```python
# 1. Điền giá trị thiếu trong cột age bằng trung bình
df['Age'].fillna(df['Age'].mean(), inplace=True)
# loại bỏ các dòng chứa giá trị thiếu
# df.dropna(inplace=True)

#  điền giá trị thiếu trong name = 'UnknownArithmeticError'
df['Name'].fillna('UnknownArithmeticError', inplace=True)
#  điền giá trị thiếu trong salary bằng phương pháp Interpolation
df['Salary'].interpolate(method='linear', inplace=True)

print(df)
```

```python
import pandas as pd
df1 = pd.DataFrame({
    "Id" : [1,2,3],
    "Score_A" : [70, 90, 85]
})

df2 = pd.DataFrame({
    "Id" : [3,4,5],
    "Score_B" : [72, 91, 75]
})

# Thực hiện merge trên cột ID (inner join, left join, outer join)
df3 = pd.merge(df1, df2, on='Id', how='inner')

df3 = pd.merge(df1, df2, on='Id', how='left')

df3 = pd.merge(df1, df2, on='Id', how='outer')
df3
# Nối dataframe theo chiều dọc
df4 = pd.concat([df1, df2], axis=0)
# df4
# Gộp df1 và df2 để điền giá trị thiếu
df5 = pd.merge(df1, df2, on='Id', how='outer').fillna(0)
df5

df6 = df1.set_index('Id').combine_first(df2.set_index('Id')).reset_index()
df6
```

```python
data = pd.DataFrame({
    "ID": range(1, 100001),
    "value": np.random.randint(0, 100, 100000)
})

print(data)
#  dùng .astype() để chuyển tối ưu hóa bộ nhớ
data['ID'] = data['ID'].astype('int32')
data['value'] = data['value'].astype('int32')

#  tìm 5 giá trị phổ biến nhất trong value
top_5_values = data['value'].value_counts().head(5)
print(top_5_values)

# Sử dụng query để lọc dữ liệu nhanh hơn của value > 90
filtered_data = data.query('value > 90')
print(filtered_data)
```

## 3. Matplotlib

```python
# Biểu đồ nhiều đường
# Cho dữ liệu sau

import matplotlib.pyplot as plt
import numpy as np  
import pandas as pd

t = np.linspace(0,10,100)
y1 = np.sin(t)
y2 = np.cos(t)
y3 = np.sin(t) * np.cos(t)


```

```python
# Tạo biểu đồ nhiều đường theo t
plt.plot(t, y1, label='sin(t)')
plt.plot(t, y2, label='scos(t)')
plt.plot(t, y3, label='sin(t) * cos(t)')
```

```python
# 2. Biểu đồ thanh nhóm
# Cho dữ liệu sau
lable = ["Q1","Q2", "Q3", "Q4" ]
A = [500,700,800,600]
B = [450, 350, 650, 750]
C = [500,250,850,600]

x = np.arange(len(lable))
width = 0.3
plt.bar(x-width, A, width, label='A')
plt.bar(x, B, width, label='B')
plt.bar(x + width, C, width, label='C')

```
